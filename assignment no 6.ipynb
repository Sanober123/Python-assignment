```python
#Q NO 2
BENIFITS OF OBJECT ORIENTED PROGRAMMING
1-The class is sharable so code is reusable.It means reusing some facilities rather than building it again and again.
2-Programming gets easier to efficient.
3-Productivity of programs increases.
4-Data is safe and secure.
```
